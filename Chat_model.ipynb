```python
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-pro")
result = llm.invoke("Write a ballad about LangChain")
print(result.content)
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-pro")
result = llm.invoke("Enter your question")
print(result.content)
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-pro")
result = llm.invoke("what is artificall intelligence")
print(result.content)
```

```python
for chunk in llm.stream("What is different between autoencoder and decoder."):
    print(chunk.content)
    print("---")
# Note that each chunk may contain more
```

```python
results = llm.batch(
    [
        "What's 2+2?",
        "What's 3+5?",
    ]
)
for res in results:
    print(res.content)
```

```python
results = llm.batch(
    [
        "What's 6/2?",
        "What's 3x5?",
    ]
)
for res in results:
    print(res.content)
```

```python
results = llm.batch(
    [
        "What's 15%3?",
        "What's 123x567?",
    ]
)
for res in results:
    print(res.content)
```
